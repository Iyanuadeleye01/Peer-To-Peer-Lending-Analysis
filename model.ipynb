# Hyperparameter tuning.

```python
from lightgbm import LGBMClassifier
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
# Initialize LightGBM classifier

lgb_model_final = LGBMClassifier(objective='binary', class_weight='balanced', boosting_type='gbdt')
```

```python
# Set the Hyperparameters for LightGBM

param_grid = {
    'num_leaves': [31, 63, 127],
    'max_depth': [5, 7,10,20],
    'learning_rate': [0.01, 0.05],
    'n_estimators': [100, 150, 200, 500],
    'colsample_bytree': [0.8, 1.0],
    'subsample': [0.8, 1.0],
    'reg_alpha': [0, 0.1],
    'reg_lambda': [0, 0.1]
}
```

```python
random_search = RandomizedSearchCV(estimator = lgb_model_final, 
                                   param_distributions = param_grid,
                                   n_iter=20, 
                                   scoring='roc_auc',   
                                   cv=5,
                                   random_state=42, 
                                   verbose=2, 
                                   n_jobs=-1
                                  )

random_search.fit(X_train_MI_upsampled, y_train_MI_upsampled)
```

```python
# Best parameters and best model
best_params = random_search.best_params_
best_model = random_search.best_estimator_

print("Best parameters found: ", best_params)
```

```python
# Train the model with the random search best parameters
best_model = lgb.LGBMClassifier(**random_search.best_params_, objective='binary', class_weight='balanced', boosting_type='gbdt')

best_model.fit(X_train_MI_upsampled, y_train_MI_upsampled)
```

```python
# Prediction with test data

y_predict_b = best_model.predict(X_test_MI)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test_MI, y_predict_b)
print(f"Accuracy: {accuracy:.4f}")

# Print classification report
print("\nClassification Report:")
print(classification_report(y_test_MI, y_predict_b))

# Print confusion matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test_MI, y_predict_b))

# Print confusion matrix
print("AUC score:")
print(roc_auc_score(y_test_MI, y_predict_b))
```
